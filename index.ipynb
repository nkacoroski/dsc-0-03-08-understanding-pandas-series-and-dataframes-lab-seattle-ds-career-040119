## Create another column 'Num_Lines' that is a count of how many lines pass through a station. Then sort your dataframe by this column in descending order.

```python
# df.groupby('station')['linename'].value_counts()

# df.groupby(['station', 'linename']).size()
    
line_counts = dict(df.groupby('station')['linename'].nunique())
df['Num_Lines'] = df.station.map(line_counts)
df = df.sort_values(by='Num_Lines', ascending=False)
df.head()
```

## Compare subway traffic by day of the week. Display this as a graph.

```python
df.date.iloc[0]
```

```python
df['date'] = pd.to_datetime(df.date, format='%m/%d/%Y')
df_1 = df
```

```python
df_1.head()
```

```python
df['weekday'] = df_1.date.dt.day_name()

df_1['total_t'] = df_1['entries'] + df_1['exits']
df_1.head()
```

```python
df_s = df_1
week_traffic = dict(df_s.groupby('weekday')['total_t'].sum())
week_exits = dict(df_s.groupby('weekday')['exits'].sum())
week_entries = dict(df_s.groupby('weekday')['entries'].sum())
week_entries
```

```python
week = {'Sunday': None, 'Monday': None, 'Tuesday': None, 'Wednesday': None, 'Thursday': None, 'Friday': None, 'Saturday': None}
week
```

```python
for key in week:
    week[key] = [round(week_traffic[str(key)] / 1e12, 2), round(week_entries[str(key)] / 1e12, 2), round(week_exits[str(key)] / 1e12, 2)]
week
# round to trillions
```

```python
x = list(week.keys())
y_1 = [week[key][0] for key in week]
y_2 = [week[key][1] for key in week]
y_3 = [week[key][2] for key in week]
 
index = np.arange(len(y_1))
bar_width = 0.8
opacity = 0.8
plt.figure(figsize=(10, 8))
y1 = plt.bar(index, y_1, bar_width, alpha=opacity, color='b', label='total traffic')
y2 = plt.bar(index, y_2, bar_width, alpha=opacity, color='r', label='entries')
y3 = plt.bar(index, y_3, bar_width, alpha=opacity, color='g', label='exits')

plt.xlabel('Weekday')
plt.ylabel('Traffic (trillion)')
plt.title('Subway Traffic by Day of the Week')
plt.xticks(index, (week.keys()))
plt.yticks(np.arange(0, 2.4, 0.2))
plt.legend((y1[0], y2[0], y3[0]), ('Total Traffic', 'Entries', 'Exits'))

plt.show()
```
